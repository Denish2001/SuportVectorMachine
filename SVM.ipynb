```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets._samples_generator import make_blobs
```

```python
#make forty different points
X, y = make_blobs(n_samples=40, centers=2, random_state=20)
```

```python
# fit the model and dont regularize
clf=svm.SVC(kernel='linear', C=1000)
clf.fit(X,y)
```

```python
#Create the graph
plt.scatter(X[:,0],X[:,1], c=y, s=30, cmap=plt.cm.Paired)
plt.show()
```

```python
newdata = [[2,3],[8,6]]
print(clf.predict(newdata))
```

```python
clf=svm.SVC(kernel='linear', C=1000)
clf.fit(X,y)
plt.scatter(X[:,0],X[:,1], c=y, s=30, cmap=plt.cm.Paired)

#plot the descission function
ax=plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

#create the grid for evaluation
xx=np.linspace(xlim[0],xlim[1],30)
yy=np.linspace(ylim[0],ylim[1],30)

YY,XX = np.meshgrid(yy,xx)
xy=np.vstack([XX.ravel(), YY.ravel()]).T
Z=clf.decision_function(xy).reshape(XX.shape)

#plot decission boundaries and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha = 0.5, linestyles=['--','-','--'])

#plot support vector
ax.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1], s=300, linewidth=1, facecolors='#A0F0CC')
plt.show()
```
